```python
#importa pandas para usarlo como pd
import pandas as pd
```

```python
#recupera el archivo PostA desde hadoop
postQ = pd.read_csv('PostQ_Filtrado.csv')
```

# Nueva sección

```python
# muestra los primeros 5 resultados del DataFrame
postQ.head(61586)
```

```python
# modifica la columna Tags como una Series de pandas para su tratamiento
postQ["Tags"] = pd.Series(postQ['Tags'], dtype="string")
```

```python
# muestrel DataFrame modificado
postQ.head(61586)
```

```python
# construye una lista con los tags separados por , cada vez que encuentra "><"
postQ["Tags"]  = postQ["Tags"].str.split("><")
```

```python
# muestra el DataFrame modificado
postQ.head(61586)
```

```python
# El index obliga a que el nuevo dataframe sea de todas las filas y no solo de 3.
postQTags = pd.DataFrame(columns=["Id","Tags"],
                        index=range(61586))
```

```python
# Recupera la columna de Post como un DataFrame de Series.
postQTags["Tags"]= pd.DataFrame(postQ["Tags"])
postQTags["Id"]= pd.DataFrame(postQ["Id"])
```

```python
postQTags.head(61586)
```

```python
#Inicializacion de DataFrames para el tratado de tags
pqt = pd.DataFrame()
postGral = pd.DataFrame()
```

```python
# Se obtiene la longitud del dataframe a tratar para corte del for
cantidadPostQ = len(postQTags.index)
```

```python
# se recorre el DataFrame de Series tags completo mediante el for
for i in range(0,cantidadPostQ):
    #se inicializa pqt cada vez para que este limpio en cada vuelta y permita el guardado nuevo
    pqt=pd.DataFrame()
    pqt["Tags"] = postQTags.loc[i,"Tags"]
    pqt["Id"] = postQTags.loc[i,"Id"]
    if i==0:
        postGral ["Tags"]= pqt ["Tags"]
        postGral ["Id"]= pqt ["Id"]
    else:
        postGral = postGral.append(pqt, ignore_index=True)
```

```python
# Resetea el indice para que quede correlativo
postGral.reset_index
```

```python
# Se limpian los caracteres de apertura y cierre de la cadena de tags origial
postGral["Tags"] = postGral.replace(regex=['<'],value='')
postGral["Tags"] = postGral.replace(regex=['>'],value='')
```

```python
# Guarda el resultado en un .csv
postGral.to_csv("PostQTagsIdFiltrado.csv")
```
