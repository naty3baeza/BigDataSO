```python
import numpy as np
# instalar sklearn --> pip3 install sklearn
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```python
pCorrSV6 = pCorrSV3[300:550]
print(pCorrSV6)
x_1 = pCorrSV6['ViewCount']
y_1 = pCorrSV6['Score']
```

```python
X = np.array(x_1).reshape(-1,1)
y = y_1
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr=linear_model.LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
#plt.figure(figsize=(10, 3))
ax = plt.axes()
ax.scatter(X, y)
ax.plot(X_test, y_pred, color = "black")

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')


plt.show()
```
