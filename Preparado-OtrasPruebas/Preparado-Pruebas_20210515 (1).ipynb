```python
import xml.etree.ElementTree as etree
xmlString= "<tags><row Id='1' TagName='.net' Count='1069' ExcerptPostId='680' WikiPostId='679' /></tags>"
xml= etree.fromstring(xmlString)  

def get_attr(xml):
    attributes = []
    for child in (xml):
        if len(child.attrib)!= 0:
            attributes.append(child.attrib)
        get_attr(child)
    return attributes

attributes = get_attr(xml)

print(attributes)
```

```python
import xml.etree.ElementTree as ET #importa la libreria que parsea xmls y la referencia como ET
import pandas as pd

TagName = []
Count = []
archivo_xml = ET.parse('Tags.xml') #parsea el archivo de Tags
raiz = archivo_xml.getroot() #obtiene el elemento raiz
for hijo in raiz: #itera sobre los hijos de raiz
    TagName.append(hijo.attrib.get('TagName'))
    Count.append(hijo.attrib.get('Count'))
df = pd.DataFrame({'Nombre Tag':TagName, 'Cantidad':Count})
df.to_csv('pruebaNueva.csv')
```

```python
import bqplot.pyplot as plt
import xml.etree.ElementTree as ET #importa la libreria que parsea xmls y la referencia como ET

TagName = []
Count = []
archivo_xml = ET.parse('Tags.xml') #parsea el archivo de Tags
raiz = archivo_xml.getroot() #obtiene el elemento raiz
for hijo in raiz: #itera sobre los hijos de raiz
    TagName.append(hijo.attrib.get('TagName'))
    Count.append(hijo.attrib.get('Count'))

fig = plt.figure(padding_x=0)
axes_options = {'x': {'label': 'X'}, 'y': {'label': 'Y'}}
plt.bar(x=TagName, y=Count, axes_options=axes_options)
fig
```
